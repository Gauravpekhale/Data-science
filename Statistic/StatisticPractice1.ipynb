```python
import numpy as np
import pandas as pd
import statistics

```

```python
# Data 
ages=[23,24,32,45,12,43,67,45,32,56,32,100]
```

```python
# Measure Of Central Tendancy
Muu=np.mean(ages)
Median=np.median(ages)
Mode=statistics.mode(ages)
print("Mean  - ",Muu)
print("Median  - ",Median)
print("Mode  - ",Mode)
```

```python
# Box Plot
import seaborn as sns
sns.boxplot(ages)
```

```python
#5 Number Summary

q1,q3=np.percentile(ages,[25,75])
IQR=q3-q1
Lower_fense=q1-1.5*IQR
Upperr_fense=q3+1.5*IQR


print("Lower _fense ",Lower_fense," <--------------> Upperr_fense ",Upperr_fense)
```

```python
# Measure of Dispersion or Spread 
variance=np.var(ages,axis=0)  
print(variance)      # Sample Variance
print(statistics.variance(ages))   # Population Variance

import math

SD=math.sqrt(variance) 
print(SD)
```

```python
#Normal Distibution 
s=np.random.normal(0.5,0.2,1000)
```

```python
#log Normal 

Mu=3
sigma=1
s=np.random.lognormal(Mu,sigma,1000)

sns.histplot(s,kde=True)
```

```python
# Coverting log normal to Normal 
sns.histplot(np.log(s),kde=True)
```

```python
#Check Wheather the data is in normal form or not ?

import matplotlib.pyplot as plt
import scipy.stats as stats
import pylab

def Plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stats.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
s=np.random.normal(0.5,0.2,10000)
Plot_data(s)
```
